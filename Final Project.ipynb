```python
data3['release_date_dummy'] = data3['release_date'].apply(lambda x: x.split('-')[0]).astype(int)
```

```python
wanted_songs = data3[data3["release_date_dummy"]>2007]
```

```python
wanted_songs['artists_dummy'] = wanted_songs['artists'].apply(lambda x: x[1:-1].split(',')).apply(lambda lis: " ".join([art.strip('"\' ') for art in lis]))
```

```python
queries = wanted_songs[['name','artists_dummy']].apply(" ".join, axis=1)
```

```python
query_list = queries.values.tolist()
```

```python
sample = query_list[-1400:-1350]
```

```python
sample
```

```python
# This costs 100 per execution. 
def search_song(song_query=''):
    song_video_id = youtube.search().list(
            part="id",
            type='video',
            regionCode="US",
            order="relevance",
            q=song_query,
            maxResults=1,
            fields="items(id(videoId))"
    ).execute()
    return song_video_id
```

```python
from tqdm import tqdm
song_ids = dict()
for song_name in tqdm(sample):
    res = search_song(song_name)
#         song_ids[song_name] = res['items'][0]['id']['videoId']
    song_ids[song_name] = res
```

```python
song_ids_cleaned = dict()
for song, id_tuple in song_ids.items():
    song_ids_cleaned[song] = id_tuple['items'][0]['id']['videoId']
print(len(song_ids_cleaned))
```

```python
BATCH_RANGE = '24256_24306'
```

```python
import json
with open(f'SongsIds_{BATCH_RANGE}.json', 'w') as fp:
    json.dump(song_ids_cleaned, fp)
```
